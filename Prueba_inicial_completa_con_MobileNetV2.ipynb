## **2. Inspección del conjunto de datos**

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
import random

# Ruta al dataset de entrenamiento
train_path = '/kaggle/input/imagenetmini-1000/imagenet-mini/train'
val_path = '/kaggle/input/imagenetmini-1000/imagenet-mini/val'

# Obtener solo carpetas (clases) válidas
clases = sorted([d for d in os.listdir(train_path) if os.path.isdir(os.path.join(train_path, d))])
print(f"[INFO]: Clases disponibles: f{len(clases)}  {clases}")

# Preparar lista de (imagen, etiqueta)
ejemplos = []
for clase in clases:
    ruta_clase = os.path.join(train_path, clase)
    imagenes = [f for f in os.listdir(ruta_clase) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    for img in imagenes:
        ejemplos.append((os.path.join(ruta_clase, img), clase))

# Seleccionar aleatoriamente 28 imágenes
muestras = random.sample(ejemplos, 28)

# Visualización tipo CIFAR-10
fig = plt.figure(figsize=(14, 10))
for i, (ruta_img, etiqueta) in enumerate(muestras):
    fig.add_subplot(4, 7, i+1)
    img = Image.open(ruta_img).resize((224, 224))
    plt.imshow(img)
    plt.title(etiqueta, fontsize=8)
    plt.axis("off")

plt.tight_layout()
plt.suptitle("Muestra aleatoria del dataset de vegetales", fontsize=16, y=1.02)
plt.show()
```

## Paso 2: Importar librerías necesarias

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.applications import ResNet50, MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import to_categorical
```

Asumiré que tenemos un dataset organizado en carpetas por clases, con estructura:

```python
'''
dataset/
    train/
        class1/
        class2/
        ...
    val/
        class1/
        class2/
        ...
    test/
        class1/
        class2/
        ...
'''
# Configuración de rutas y parámetros
'''
DATA_DIR = 'imagenet-mini'  # Cambiar por la ruta real
TRAIN_DIR = os.path.join(DATA_DIR, 'train')
VAL_DIR = os.path.join(DATA_DIR, 'val')
TEST_DIR = os.path.join(DATA_DIR, 'test')
'''
TRAIN_DIR = '/kaggle/input/imagenetmini-1000/imagenet-mini/train'
VAL_DIR = '/kaggle/input/imagenetmini-1000/imagenet-mini/val'
TEST_DIR = '/kaggle/input/imagenetmini-1000/imagenet-mini/val'
IMG_SIZE = (224, 224)  # Tamaño para ResNet50 y MobileNetV2
BATCH_SIZE = 32
NUM_CLASSES = 1000  # Para ImageNetMini-1000

# Generadores de datos con aumento para entrenamiento
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Flujos de datos
train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    VAL_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    #TEST_DIR,
    VAL_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

```

```python
# Arquitectura de red - modo API Funcional
from tensorflow.keras.models import Model
def build_resnet50_model(num_classes):
    # Cargar ResNet50 preentrenada (excluyendo las capas fully connected)
    base_model = ResNet50(
        weights='imagenet',
        include_top=False,
        input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3)
    )
    '''
    # Capa de entrada
    inputs = layers.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))

    # Cargar ResNet50 preentrenada
    x = base_model(inputs, training=False)

    #1 BASE MODEL

    # Bloque 1
    x1 = layers.Conv2D(32, (3, 3), padding='same', activation="relu")(x)
    x1 = layers.BatchNormalization()(x1)
    x1 = layers.Dropout(0.1)(x1)
    x1 = layers.Conv2D(32, (3, 3), padding='same', activation="relu")(x1)
    x1 = layers.BatchNormalization()(x1)
    x1 = layers.Dropout(0.1)(x1)
    x1 = layers.MaxPooling2D((1, 1))(x1)

    # Bloque 2
    x2 = layers.Conv2D(64, (3, 3), padding='same', activation="relu")(x1)
    x2 = layers.BatchNormalization()(x2)
    x2 = layers.Dropout(0.1)(x2)
    x2 = layers.Conv2D(64, (3, 3), padding='same', activation="relu")(x2)
    x2 = layers.BatchNormalization()(x2)
    x2 = layers.Dropout(0.1)(x2)
    x2 = layers.MaxPooling2D((1, 1))(x2)

    # Bloque 3
    x3 = layers.Conv2D(256, (3, 3), padding='same', activation="relu")(x2)
    x3 = layers.BatchNormalization()(x3)
    x3 = layers.Dropout(0.1)(x3)
    x3 = layers.Conv2D(256, (3, 3), padding='same', activation="relu")(x3)
    x3 = layers.BatchNormalization()(x3)
    x3 = layers.Dropout(0.1)(x3)
    x3 = layers.MaxPooling2D((1, 1))(x3)

    # 2. TOP MODEL

    # Tendencias actuales:
    xfc = layers.GlobalAveragePooling2D()(x3)
    predictions = layers.Dense(num_classes, activation='softmax')(xfc)
    #model = tf.keras.Model(inputs, predictions)

    # 3. Unimos las entradas y salidas del modelo
    convnet = Model(inputs=inputs, outputs=predictions)
    convnet.summary()

    # Compilación
    convnet.compile(optimizer=optimizers.Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return convnet
    '''

    # Congelar las capas del modelo base
    base_model.trainable = False

    # Construir el modelo completo
    inputs = tf.keras.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
    x = base_model(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dense(512, activation='relu')(x)
    #x = layers.Dense(512, activation='swish',kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)
    x = layers.Dropout(0.1)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dense(256, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(0.001))(x)
    x = layers.Dropout(0.1)(x)
    outputs = layers.Dense(num_classes, activation='softmax',kernel_initializer='glorot_uniform')(x)

    model = tf.keras.Model(inputs, outputs)

    # Compilar el modelo
    model.compile(
        optimizer=optimizers.Adam(learning_rate=1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

def build_mobilenetv2_model(num_classes):
    # Cargar MobileNetV2 preentrenada
    base_model = MobileNetV2(
        weights='imagenet',
        include_top=False,
        input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3)
    )

    # Congelar las capas del modelo base
    base_model.trainable = False

    # Construir el modelo completo
    inputs = tf.keras.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
    x = base_model(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(1024, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = tf.keras.Model(inputs, outputs)

    # Compilar el modelo
    model.compile(
        optimizer=optimizers.Adam(learning_rate=1e-3),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

## Paso 5: Entrenamiento con Transfer Learning
Entrenaremos ambos modelos usando solo las nuevas capas añadidas, manteniendo congeladas las capas preentrenadas

```python
# Callbacks comunes para ambos modelos
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)
model_checkpoint = ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_accuracy')
'''
# Entrenar ResNet50
print("\nEntrenando ResNet50 con Transfer Learning...")
resnet_model = build_resnet50_model(NUM_CLASSES)
resnet_history = resnet_model.fit(
    train_generator,
    epochs=15,
    validation_data=val_generator,
    callbacks=[early_stop, reduce_lr, model_checkpoint]
)
'''
# Entrenar MobileNetV2
print("\nEntrenando MobileNetV2 con Transfer Learning...")
mobilenet_model = build_mobilenetv2_model(NUM_CLASSES)
mobilenet_history = mobilenet_model.fit(
    train_generator,
    epochs=20,
    validation_data=val_generator,
    callbacks=[early_stop, reduce_lr, model_checkpoint]
)
```

## Paso 6: Fine-Tuning
Después del entrenamiento inicial, haremos fine-tuning descongelando algunas capas del modelo base.

```python
def fine_tune_model(model, base_model):
    # Descongelar las últimas capas del modelo base
    base_model.trainable = True
    for layer in base_model.layers[:-10]:
        layer.trainable = False

    # Recompilar el modelo con una tasa de aprendizaje más baja
    model.compile(
        optimizer=optimizers.Adam(learning_rate=1e-5),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
'''
# Fine-tuning para ResNet50
print("\nFine-tuning ResNet50...")
resnet_model = fine_tune_model(resnet_model, resnet_model.layers[1])
resnet_ft_history = resnet_model.fit(
    train_generator,
    epochs=15,
    initial_epoch=resnet_history.epoch[-1],
    validation_data=val_generator,
    callbacks=[early_stop, reduce_lr, model_checkpoint]
)
'''
# Fine-tuning para MobileNetV2
print("\nFine-tuning MobileNetV2...")
mobilenet_model = fine_tune_model(mobilenet_model, mobilenet_model.layers[1])
mobilenet_ft_history = mobilenet_model.fit(
    train_generator,
    epochs=15,
    initial_epoch=mobilenet_history.epoch[-1],
    validation_data=val_generator,
    callbacks=[early_stop, reduce_lr, model_checkpoint]
)
```

## Paso 7: Evaluación y comparación de modelos
Evaluaremos ambos modelos en el conjunto de prueba y compararemos su rendimiento.

```python
def evaluate_model(model, test_generator, model_name):
    print(f"\nEvaluando {model_name} en el conjunto de prueba...")
    test_loss, test_acc = model.evaluate(test_generator)
    print(f"Precisión en test: {test_acc:.4f}")
    print(f"Pérdida en test: {test_loss:.4f}")

    # Predicciones
    y_pred = model.predict(test_generator)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true = test_generator.classes

    # Reporte de clasificación
    print("\nReporte de clasificación:")
    print(classification_report(y_true, y_pred_classes, target_names=test_generator.class_indices.keys()))

    return test_acc, test_loss

# Evaluar ambos modelos
#resnet_acc, resnet_loss = evaluate_model(resnet_model, test_generator, "ResNet50")
mobilenet_acc, mobilenet_loss = evaluate_model(mobilenet_model, test_generator, "MobileNetV2")

# Comparación final
print("\nComparación final:")
#print(f"ResNet50 - Precisión: {resnet_acc:.4f}, Pérdida: {resnet_loss:.4f}")
print(f"MobileNetV2 - Precisión: {mobilenet_acc:.4f}, Pérdida: {mobilenet_loss:.4f}")
```

## Paso 8: Visualización de resultados
